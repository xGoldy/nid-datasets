This notebook briefly presents the structure of a CSV AWID3 dataset format. However, PCAP format is available as well.

```python
import pandas as pd
import numpy as np
import os

pd.set_option('display.max_columns', None)
```

```python
DATAPATH = '/data/disk2/awid3/CSV/'
```

```python
# List folder contents
for root, dirs, files in os.walk(DATAPATH):
    if not os.path.basename(root) == "":
        print(f'{os.path.basename(root)}/')
    for file in sorted(files):
        print('   ', file)
```

As apparent, there are several folders split by the attack type with a lot of files. We will count them...

```python
dirs_total = 0
files_total = 0

for root, dirs, files in os.walk(DATAPATH):
    if not os.path.basename(root) == "":
        dirs_total += 1

    files_total += len(list(files))

dirs_total, files_total
```

So there are 13 different attack types with 745 files in total. Count the total number of packets.

```python
# Count the total number of packets
!( find $DATAPATH -name '*.csv' -print0 | xargs -0 cat ) | wc -l
```
