Just based on this, we see that the data are not sorted by timestamps of the start. They still may be sorted by the end however.

The algorithm below could be more easily tackled by Big-Data libraries like Dask or Polars. However, I somehow forgot they exist and decided to code an own algorithm for the purpose instead.

```python
# Iterate through the whole dataframe and save classes value counts
# and length of the df + measure if the entries are sorted
import datetime

data_all = pd.read_csv(DATAPATH, chunksize=1000000)
all_classes_counts_a = pd.Series(dtype=int)
all_classes_counts_t = pd.Series(dtype=int)
total_len = 0
last_overall_timestamp = None
timestamps_sorted_ascending = True
timestamps_sorted_descending = True

for chunk in data_all:
    chunk_a_counts = chunk['attack_a'].value_counts()
    chunk_t_counts = chunk['attack_t'].value_counts()
    last_entry = chunk.iloc[-1]

    all_classes_counts_a = all_classes_counts_a.add(chunk_a_counts, fill_value=0)
    all_classes_counts_t = all_classes_counts_t.add(chunk_t_counts, fill_value=0)

    total_len += len(chunk)

    last_chunk_timestamp = datetime.datetime(
        year=last_entry['te_year'],
        month=last_entry['te_month'],
        day=last_entry['te_day'],
        hour=last_entry['te_hour'],
        minute=last_entry['te_min'],
        second=last_entry['te_second']
    )

    # Optimized chunk-check for ascending timestamps
    if last_overall_timestamp is not None and timestamps_sorted_descending:
        if last_overall_timestamp > last_chunk_timestamp:
            timestamps_sorted_descending = False

            print("Descending condition broken:")
            print(f"Current: {last_chunk_timestamp}\nOverall: {last_overall_timestamp}")

    # Optimized chunk-check for descending timestamps
    if last_overall_timestamp is not None and timestamps_sorted_ascending:
        if last_chunk_timestamp > last_overall_timestamp:
            timestamps_sorted_ascending = False

            print("Ascending condition broken:")
            print(f"Current: {last_chunk_timestamp}\nOverall: {last_overall_timestamp}")

    last_overall_timestamp = last_chunk_timestamp


all_classes_counts_a = all_classes_counts_a.astype(np.int32)
all_classes_counts_t = all_classes_counts_t.astype(np.int32)
```
