```python
# Timestamps analysis
timestamps = data['timestamp']
timestamps.head()
```

```python
# Convert the timestamps into the Datetime format
timestamps = pd.to_datetime(timestamps, format='%d/%b/%Y:%H:%M:%S %z', utc=True)
timestamps
```

```python
# Sort the timestamps to determine the span of the capture
timestamps = timestamps.sort_values(ascending=False).dropna()
timestamps = timestamps.reset_index(drop=True)
timestamps
```

```python
# Display the span of the capture (start, end)
span = timestamps.iloc[0] - timestamps.iloc[-1]
span
```

```python
# How big is the 1% span for this?
span * 0.01
```

That is way too much for this case, consider 1 hour at most when measuring the duration, the data were discontinous anyways even when this span is used.

```python
# Are there gaps in the data (was the capture interrupted?)
def measure_real_capture_dur(data: pd.Series, gap_max_secs: int = 300) -> float:
    """Computes total timespan of the capture. Expects iterable containing timestamps objects sorted in a descending manner"""
    total_dur = pd.Timedelta(seconds=0)
    current_dur = pd.Timedelta(seconds=0)
    cont_durations = []
    last_tstamp = data.iloc[0]
    contiguous = True

    # Iterate through the dataframe to find out gaps
    for cur_tstamp in data:
        dur_gap = last_tstamp - cur_tstamp

        if dur_gap <= pd.Timedelta(seconds=gap_max_secs):
            total_dur += dur_gap
            current_dur += dur_gap
        else:
            cont_durations.append(current_dur)
            current_dur = pd.Timedelta(seconds=0)

            if contiguous:
                contiguous = False

        last_tstamp = cur_tstamp

    # Log the final continuous block and sort them via their length
    cont_durations.append(current_dur)
    cont_durations.sort(reverse=True)

    return total_dur, contiguous, cont_durations
```

```python
measure_real_capture_dur(timestamps, 3600)
```

Apparently, 7 days of the capture are contiguous but then everything messes up. We condiser the capture discontiguous.
