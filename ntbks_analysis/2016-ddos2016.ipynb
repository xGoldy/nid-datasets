```python
import pandas as pd

pd.set_option('display.max_columns', None)
```

```python
DDOS16_HEADER = [
    'SRC_ADD',
    'DES_ADD',
    'PKT_ID',
    'FROM_NODE',
    'TO_NODE',
    'PKT_TYPE',
    'PKT_SIZE',
    'FLAGS',
    'FID','SEQ_NUMBER',
    'NUMBER_OF_PKT',
    'NUMBER_OF_BYTE',
    'NODE_NAME_FROM',
    'NODE_NAME_TO',
    'PKT_IN',
    'PKT_OUT',
    'PKT_R',
    'PKT_DELAY_NODE',
    'PKT_RATE',
    'BYTE_RATE',
    'PKT_AVG_SIZE',
    'UTILIZATION',
    'PKT_DELAY',
    'PKT_SEND_TIME',
    'PKT_RESEVED_TIME',
    'FIRST_PKT_SENT',
    'LAST_PKT_RESEVED',
    'PKT_CLASS'
]
```

```python
data = pd.read_csv(DATAPATH, names=DDOS16_HEADER)
```
