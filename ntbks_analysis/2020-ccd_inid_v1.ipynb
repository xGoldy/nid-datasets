## Time Continousness and Duration Analysis

```python
# Time column in the CCD-INID-V1 dataset are the `bidirectional_{first|last}_seen_ms` and seen timestamps.
data['bidirectional_first_seen_ms']
```

```python
data_ts_sorted = data['bidirectional_first_seen_ms'].sort_values(ascending=False)
data_ts_sorted = pd.to_datetime(data_ts_sorted, unit='ms', origin='unix')
data_ts_sorted
```

```python
data_span = data_ts_sorted.iloc[0] - data_ts_sorted.iloc[-1]
data_span
```

```python
# Are there gaps in the data (was the capture interrupted?)
def measure_real_capture_dur(data: pd.Series, gap_max_secs: int = 300) -> float:
    """Computes total timespan of the capture. Expects iterable containing timestamps objects sorted in a descending manner"""
    total_dur = pd.Timedelta(seconds=0)
    current_dur = pd.Timedelta(seconds=0)
    cont_durations = []
    last_tstamp = data.iloc[0]
    contiguous = True

    # Iterate through the dataframe to find out gaps
    for cur_tstamp in data:
        dur_gap = last_tstamp - cur_tstamp

        if dur_gap <= pd.Timedelta(seconds=gap_max_secs):
            total_dur += dur_gap
            current_dur += dur_gap
        else:
            cont_durations.append(current_dur)
            current_dur = pd.Timedelta(seconds=0)

            if contiguous:
                contiguous = False

        last_tstamp = cur_tstamp

    # Log the final continuous block and sort them via their length
    cont_durations.append(current_dur)
    cont_durations.sort(reverse=True)

    return total_dur, contiguous, cont_durations
```

```python
# How big is the tolerance of 1% maximum gap?
data_span * 0.01
```

```python
measure_real_capture_dur(data_ts_sorted, int(data_span.total_seconds() * 0.01))
```

```python
# How long is the real capture with much lower tolerance -- e.g. 1200 sec - 20 min
measure_real_capture_dur(data_ts_sorted, 1200)
```
