```python
# Data path
#DATAPATH = '/data/kinit/gold_data/cse-cic-ids2018/dataset/csv'
DATAPATH = '/data/datasets_cic/cse_cic2018'
```

```python
import pandas as pd
import numpy as np
import os

pd.set_option('display.max_columns', None)
```

```python
# See the files in the directory
data_files = os.listdir(DATAPATH)
data_files
```

```python
# Read all the days within a single file
data_all = pd.DataFrame()

for file in data_files:
    fullpath = os.path.join(DATAPATH, file)

    print(file)
    data_all = pd.concat([data_all, pd.read_csv(fullpath)])
```
